### 2)	List all book's book_id published by publisher Publisher1.

```python
def books_ji(temp1):
    pub_lis=[]
    for i in publishers:
        if(i.name == temp1):
            temp=i.publisher_id
            break
    for j in books:
        if(j.publisher_id==temp):
            pub_lis.append(j.book_id)
    pub_lis=list(set(pub_lis))
    for x in pub_lis:
        print("Book_id published by publisher1 is :",x)
    return pub_lis
```

### 3)	find the author_id of author Jane Doe, where Jane is the author's first name and Doe is her last name.

```python
def auth(fname,lname):
    for i in authors:
        if(i.first_name==fname and i.last_name==lname):
            print("author_id of jane doe is :", i.author_id)
            temp=i.author_id
    return temp
```

```python
auth("Jane","Doe")
```

### 4)	(use the result from question 3), find all books' book_id authored by Jane Doe.

```python
def jdbooks(fname,lname):
    tem=auth(fname,lname)
    k=[]
    for i in book_authors:
        if(tem==i.author_id):
            print("Book's book_id authored by Jane Doe is : ",i.book_id)
            k.append(i.book_id)
    return k
```

```python
jdbooks("Jane","Doe")
```

### 5)	(use the result from question 3 and 4), find Jane Doe's coauthors' first_name and last _name. coauthors can be found by writing a same book but with different author_id. 

```python
def coauthors(fname,lname):
    aid=auth(fname,lname)
    bid=jdbooks(fname,lname)
    tmp=[]
    for i in book_authors:
        if(i.book_id in bid and i.author_id not in [aid]):
            for j in authors:
                if(j.author_id==i.author_id):
                    tmp.append(j.first_name+"  "+ j.last_name)
    tmp=list(set(tmp))
    for v in tmp:
        print("CoAuthors first_name and last-name is:",v)
                        
    return None
```

```python
coauthors("Jane","Doe")
```

### 6)	find author's author_id who has published more than one book.

```python
def authorbooks():
    tmp=[]
    for i in book_authors:
        for j in book_authors:
            if(i.author_id == j.author_id and i.book_id != j.book_id):
                tmp.append(i.author_id)
    tmp=set(tmp)
    for i in tmp:
        print("author's author_id who printed more than one book is : ", i)
    return None
        
```
